### OUTLIERS REMOVAL

## ZSCORE

```python
from scipy.stats import zscore

z= np.abs(zscore(ic.select_dtypes(include= ['int64', 'float', 'int32'])))
print(np.where(z>3))
```

```python
ic_1= ic[(z<3).all(axis=1)]
print("With Outliers:", ic.shape)
print("After Removing Outliers:", ic_1.shape)
```

## IQR

```python
# Using IQR Method

from scipy import stats
IQR= stats.iqr(ic.select_dtypes(include=['int64', 'float', 'int32']))
IQR
```

```python
Q1= ic.quantile(0.25)
Q3= ic.quantile(0.75)
ic_out= ic.select_dtypes(include= ['int64', 'float', 'int32'])[~((ic.select_dtypes(include= ['int64', 'float', 'int32']) < (Q1-1.5*IQR)) | (ic.select_dtypes(include= ['int64', 'float', 'int32']) > (Q3+1.5*IQR))).any(axis=1)]
print(ic_out.shape)
```

Choosing ZScore because in IQR the dataloss is visibly more.

```python
ic= ic_1
```

### Dividing data into x and y

```python
x= ic.drop(['fraud_reported'], axis=1)
y= ic['fraud_reported']
```

```python
x.shape
```

```python
y.shape
```

### SCALING X VALUES

```python
from sklearn.preprocessing import MinMaxScaler

mm= MinMaxScaler()
x= mm.fit_transform(x)
```

```python
pd.DataFrame(x).describe()
```

Total data have been scaled.

### Using Over_Sampling_BorderlineSMOTE

```python
sns.set_style("whitegrid")
sns.countplot(x= "fraud_reported", data=ic)
```

```python
from imblearn.over_sampling import BorderlineSMOTE

x_rus, y_rus= BorderlineSMOTE().fit_resample(x,y)
print('Original Target Dataset Shape:', y.shape)
print('Resamplt Target Dataset Shape:', y_rus.shape)
```

```python
sns.set_style("whitegrid")
sns.countplot(y_rus)
```

Now we have balanced the data before model building.

### Spliting Train And Test Data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test= train_test_split(x_rus, y_rus, test_size= .27, random_state= 42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

## MODEL BUILDING

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.ensemble import GradientBoostingClassifier, AdaBoostClassifier, RandomForestClassifier, BaggingClassifier
from sklearn.naive_bayes import GaussianNB as NB
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
```

## Logistic Regression

```python
lr= LogisticRegression()
lr_score= cross_val_score(lr, x_rus, y_rus, cv=5, scoring= 'accuracy')
print("Cross Validation Score:", np.mean(lr_score))
```

```python
lr.fit(x_train, y_train)
LR_predicted= lr.predict(x_test)

print(accuracy_score(y_test, LR_predicted))
print(confusion_matrix(y_test, LR_predicted))
print(classification_report(y_test, LR_predicted))
print("Training Accuracy->", lr.score(x_train, y_train))
print("Test Accuracy->", lr.score(x_test, y_test))
```

## KNeighbors Classifier

```python
knn= KNN()
knn_score= cross_val_score(knn, x_rus, y_rus, cv=5, scoring= 'accuracy')
print("Cross Validation Score:", np.mean(knn_score))
```

```python
knn.fit(x_train, y_train)
Knn_predicted= knn.predict(x_test)

print(accuracy_score(y_test, Knn_predicted))
print(confusion_matrix(y_test, Knn_predicted))
print(classification_report(y_test, Knn_predicted))
print("Training Accuracy->", knn.score(x_train, y_train))
print("Test Accuracy->", knn.score(x_test, y_test))
```

## HYPER PARAMETER TUNING

### GRADIENT BOOSTING CLASSIFIER

```python
Gbc= GradientBoostingClassifier()
params_grid= {'criterion' : ['mse', 'mae'],
             'n_estimators':[100,200],
             'learning_rate':[0.1,0.5,1.0],
             'random_state':[5]}
```

```python
Gbc_grid= GridSearchCV(GradientBoostingClassifier(), params_grid, cv=4, scoring= 'accuracy', n_jobs= -1, verbose=1)
```

```python
# Accuracy after parameter tuning

Gbc_grid.fit(x_train, y_train)
Gbc_pred= Gbc_grid.best_estimator_.predict(x_test)
print("Accuracy After Parameter Tuning->", accuracy_score(y_test, Gbc_pred))
```

```python
Gbc_grid.best_params_
```

### Model Training with Best Parameters

```python
best_param= {'criterion': ['mse'],
 'learning_rate': [0.5],
 'n_estimators': [100],
 'random_state': [5]}
```

```python
best_Gbc_grid= GridSearchCV(GradientBoostingClassifier(), best_param, cv=4, scoring= 'accuracy', n_jobs= -1, verbose=2)
```

```python
best_Gbc_grid.fit(x_train, y_train)
best_Gbc_pred= best_Gbc_grid.best_estimator_.predict(x_test)
print("Accuracy After Parameter Tuning->", accuracy_score(y_test, best_Gbc_pred))
```
